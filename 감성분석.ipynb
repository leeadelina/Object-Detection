```python
import pandas as pd
import numpy as np

# 감성사전 로드
def load_sentiment_dictionary(filename):
    sentiment_dict = {}
    with open(filename, 'r', encoding='utf-8') as f:
        lines = f.readlines()
        for line in lines:
            if '\t' in line: # 탭 문자가 포함된 줄만 처리
                word, score_str = line.strip().split('\t')
                try:
                    score = float(score_str)
                    sentiment_dict[word] = score
                except ValueError:
                    print(f"점수 '{score_str}'를 float 형태로 변환할 수 없습니다. 해당 줄을 건너뜁니다.")
    return sentiment_dict

# 감성 점수 계산 함수
def calculate_sentiment_score(text, sentiment_dict):
    if pd.isna(text):  # NaN 처리
        return np.nan
    score = 0
    for word in text.split():  # 텍스트를 공백으로 분할
        if word in sentiment_dict:  # 감성 사전에 있는 단어의 감성 점수 가져오기
            score += sentiment_dict[word]
    return score
    # 감성 사전 로드
sentiment_dict = load_sentiment_dictionary('/content/drive/MyDrive/2조/감성분석_1/감성사전.txt')

# 감성 점수 열 추가
df['감성점수'] = df['캐릭터_말'].apply(lambda x: calculate_sentiment_score(x, sentiment_dict))

# 결과 출력
print(df[['캐릭터_말', '감성점수']])
```
